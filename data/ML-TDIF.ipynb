```python
import spacy
from sklearn.base import TransformerMixin

# Load the French SpaCy model
nlp = spacy.load("fr_core_news_sm")


# Define a custom transformer for lemmatization and stop words removal
class TextPreprocessor(TransformerMixin):
    def transform(self, X, **transform_params):
        return [self.preprocess(text) for text in X]

    def fit(self, X, y=None, **fit_params):
        return self

    def preprocess(self, text):
        doc = nlp(text)
        return " ".join(
            [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
        )


# Add the preprocessor to the existing pipelines
pipeline_lr.steps.insert(0, ("preprocessor", TextPreprocessor()))
pipeline_rf.steps.insert(0, ("preprocessor", TextPreprocessor()))
```
